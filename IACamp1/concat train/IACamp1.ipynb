le variable weights leur valeur centient 2 variable, le net et le gros weights

```python
t2[['gross_weight','net_weight']]=(
    t2['weights'].astype(str).str.split(',',n=1,expand=True)
)
```

```python
t2['gross_weight']=(t2['gross_weight'].astype(str).str.extract(r'(-?\d+(?:\.\d+)?)').astype(float))
```

```python
t2['gross_weight'].isna().sum()
```

```python
t2['net_weight']=(t2['net_weight'].astype(str).str.extract(r'(-?\d+(?:\.\d+)?)').astype(float))
```

```python
t2['net_weight'].isna().sum()
```

```python
t2.drop(columns='weights',inplace=True)
```
